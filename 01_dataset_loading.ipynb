#### **Download Dataset from Kaggle**

```python
import kagglehub
import os
import pandas as pd
import shutil
```

```python
# Download latest version

path = kagglehub.dataset_download("yasserh/housing-prices-dataset")

print("Path to dataset files:", path)
```

```python
# Display dataset directory path and inspect downloaded files
# This step helps identify the exact CSV filename before loading it into pandas

print("Path:", path)
print("Files inside folder:")
print(os.listdir(path))
```

```python
# Check dataset

file_path = os.path.join(path, "Housing.csv")
df = pd.read_csv(file_path)

df.head()
```

```python
# Create data/raw folder inside repository

os.makedirs("data/raw", exist_ok=True)
```

```python
# Move all files from kagglehub cache to repo folder
for file_name in os.listdir(path):
    source_file = os.path.join(path, file_name)
    destination_file = os.path.join("data/raw", file_name)
    shutil.copy(source_file, destination_file)

print("Dataset successfully saved to data/raw/")
```
