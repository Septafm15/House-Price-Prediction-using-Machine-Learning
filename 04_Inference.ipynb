# **Machine Learning Model Inference**

```python
# Library Import

import pickle
import pandas as pd
```

```python
# Load trained pipeline
with open("tuned_lasso_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)

print("Model loaded successfully")
```

```python
# Dataframe for predictions

new_houses = pd.DataFrame([
    # House 1 - Small & basic (expected lower price)
    {
        "area": 2500,
        "bedrooms": 2,
        "bathrooms": 1,
        "stories": 1,
        "mainroad": "no",
        "guestroom": "no",
        "basement": "no",
        "hotwaterheating": "no",
        "airconditioning": "no",
        "parking": 0,
        "prefarea": "no",
        "furnishingstatus": "unfurnished"
    },
    
    # House 2 - Medium standard house
    {
        "area": 6000,
        "bedrooms": 3,
        "bathrooms": 2,
        "stories": 2,
        "mainroad": "yes",
        "guestroom": "no",
        "basement": "no",
        "hotwaterheating": "no",
        "airconditioning": "yes",
        "parking": 1,
        "prefarea": "yes",
        "furnishingstatus": "semi-furnished"
    },
    
    # House 3 - Large premium house (expected higher price)
    {
        "area": 10000,
        "bedrooms": 5,
        "bathrooms": 3,
        "stories": 3,
        "mainroad": "yes",
        "guestroom": "yes",
        "basement": "yes",
        "hotwaterheating": "yes",
        "airconditioning": "yes",
        "parking": 3,
        "prefarea": "yes",
        "furnishingstatus": "furnished"
    }
])
```

```python
# Predict House Price

predictions = loaded_model.predict(new_houses)

for i, price in enumerate(predictions):
    print(f"House {i+1} Predicted Price: {price:,.0f}")
```
