```python
import pickle
import pandas as pd
```

```python
# Load trained pipeline
with open("tuned_lasso_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)

print("Model loaded successfully")
```

```python
# Dataframe for predict

new_house = pd.DataFrame([{
    "area": 6000,
    "bedrooms": 3,
    "bathrooms": 2,
    "stories": 2,
    "mainroad": "yes",
    "guestroom": "no",
    "basement": "no",
    "hotwaterheating": "no",
    "airconditioning": "yes",
    "parking": 1,
    "prefarea": "yes",
    "furnishingstatus": "semi-furnished"
}])
```

```python
prediction = loaded_model.predict(new_house)

print(f"Predicted House Price: {prediction[0]:,.0f}")
```
